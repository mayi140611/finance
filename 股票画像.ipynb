## 行情数据

```python
from pyecharts import Line
```

```python
def t(code, start_date,line):
    df = pro.daily(ts_code=code, start_date=start_date)
    #按照日期进行升序排列
    df1 = df.sort_index(ascending=False)
    line.add(
        code,
        df1['trade_date'],df1['close'],
        legend_top='bottom',
        is_more_utils=True,
        mark_line=["average"],#绘制出辅助平均线
        mark_point=[
            'min','max',
        ],
        mark_point_symbolsize=80
    )
```

```python
line = Line("行情观察",title_pos='center',subtitle="练习pyecharts使用")
```

```python
t('000001.SZ','20180701',line)
t('600029.sh','20180701',line)
```
